```python
#build dataframe for export with biological information

neuromere = "T1R"

meta_df = df_4b_t1_rhs.copy()
meta_df.head()


#define manual assignments
manual_assignments = {
    #IR
    "ir_1_1_ids": ir_1_1_ids,
    "ir_1_2_ids": ir_1_2_ids,
    "ir_1_3_ids": ir_1_3_ids,
    "ir_1_4_ids": ir_1_4_ids,
    "ir_1_5_ids": ir_1_5_ids,

    "ir_2_04B_1_ids": ir_2_04B_1_ids,
    "ir_2_04B_2_ids": ir_2_04B_2_ids,
    "ir_2_04B_3_ids": ir_2_04B_3_ids,
    "ir_2_04B_4_ids": ir_2_04B_4_ids,
    "ir_2_04B_5_ids": ir_2_04B_5_ids,
    "ir_2_04B_6_ids": ir_2_04B_6_ids,

  

    #other subclass groups
    "br_ids": br_ids,
    "bi_ids": bi_ids,
    "ia_ids": ia_ids,
    "ii_ids": ii_ids,
    "ba_ids": ba_ids,

}



# define how to interpret each group name

import re

import re

def classify_group_var(group_var):
    # IR branch 1 -> legacy tag 04B
    m1 = re.match(r"ir_1_(\d+)_ids$", group_var)
    if m1:
        return "IR", 1, int(m1.group(1)), "04B"

    # IR branch 2 -> legacy tag 20A
    m2 = re.match(r"ir_2_04B_(\d+)_ids$", group_var)
    if m2:
        return "IR", 2, int(m2.group(1)), "20A"

    # other major classes for now
    if group_var == "br_ids":
        return "BR", 1, 1, None

    if group_var == "bi_ids":
        return "BI", 1, 1, None

    if group_var == "ia_ids":
        return "IA", 1, 1, None

    if group_var == "ii_ids":
        return "II", 1, 1, None

    if group_var == "ba_ids":
        return "BA", 1, 1, None

    return None, None, None, None



# make row-wise assignment dataframe
rows = []

for var_name, body_ids in manual_assignments.items():
    if not isinstance(body_ids, list):
        continue

    major_class, subclass_level1, morphology_cluster, legacy_tag = classify_group_var(var_name)

    if major_class is None:
        continue

    morphology_cluster_id = f"{major_class}_{subclass_level1}_{morphology_cluster}"

    for body_id in body_ids:
        rows.append({
            "neuromere": neuromere,
            "bodyId": int(body_id),
            "group_var": var_name,
            "major_class": major_class,
            "subclass_level1": subclass_level1,
            "morphology_cluster": morphology_cluster,
            "morphology_cluster_id": morphology_cluster_id,
            "legacy_tag": legacy_tag,
        })

assignment_df = pd.DataFrame(rows)
assignment_df.head()
assignment_df["major_class"].unique()
```

```python
#merge metadata


final_neuromere_df = assignment_df.merge(
    meta_df[
        [
            "bodyId",
            "type",
            "instance",
            "class",
            "subclass",
            "hemilineage",
            "somaNeuromere",
            "somaSide",
            "rootSide",
            "longTract",
            "birthtime",
        ]
    ],
    on="bodyId",
    how="left"
)

print(final_neuromere_df.shape)
final_neuromere_df.head()

# a few checks:
final_neuromere_df.groupby(["legacy_tag", "morphology_cluster"])["bodyId"].nunique()

#check for duplicate assignments
dup_check = (
    final_neuromere_df.groupby(["bodyId"])["group_var"]
    .nunique()
    .reset_index(name="n_groups")
)

dup_check[dup_check["n_groups"] > 1]


#vizualize and export;
final_neuromere_df.head()



final_table_name = f"Classification_04B_{neuromere}"
print(final_table_name)

out_file = OUTPUT_DIR / f"{final_table_name}.csv"
final_neuromere_df.to_csv(out_file, index=False)

print("Saved:", out_file)

second_out_file = SECOND_OUTPUT_DIR / f"{final_table_name}.csv"
final_neuromere_df.to_csv(second_out_file, index=False)

print("Also saved:", second_out_file)
```
