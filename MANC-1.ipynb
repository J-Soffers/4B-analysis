```python
# This cell pulls out every segmentation layer’s name, segmentQuery, and the list of segment IDs it contains:
layers = state.get("layers", [])

seg_layers = [L for L in layers if L.get("type") == "segmentation"]

summary = []
for L in seg_layers:
    name = L.get("name")
    query = L.get("segmentQuery")
    segs = L.get("segments", [])
    summary.append((name, query, len(segs)))

summary_sorted = sorted(summary, key=lambda x: x[2], reverse=True)
for name, query, n in summary_sorted[:15]:
    print(f"{n:5d} segments | {name} | query={query}")
```
